```python
class TreeNode(object):
    #paraent
    #left
    #right
    #key
    #color
    def __init__(self,data,color = None,parent=None,left=None,right=None):
        self.key =data
        self.p =parent
        self.left = left
        self.right = right
        self.color = color
        
```

```python
class RB_Tree(object):
    
    def __init__(self):

        #叶节点全是黑色，无数据
        self.leaf = TreeNode(None,'Black')
         #根节点
        self.root =  self.leaf   
    def left_rotate(self,x):
        #左旋
        
        y = x.right
        
        #第一个改变
        x.right = y.left
        if y.left!=self.leaf:
            y.left.p = x
            
        #第二个改变
        if y!=self.leaf:
            y.p=x.p
        if x.p==self.leaf:
            #处理根节点情况
            self.root = y
        elif x ==x.p.left:
            x.p.left =y
        else:
            x.p.right = y
        
        #第三个改变
        y.left =x
        if x!=self.leaf:
            x.p = y
    def right_rotate(self,x):
        #右旋
        
        y = x.left
        
        #第一个改变
        x.left = y.right
        if y.right!=self.leaf:
            y.right.p = x
        
        #第二个改变
        y.p = x.p
        if x.p==self.leaf:
            self.root= y
        elif x.p.left ==x:
            x.p.left =y
        else:
            x.p.right=y
         #第三个改变
        y.right = x
        if x!=self.leaf:
            x.p = y
        
    def insert(self,value):
        #构造一个节点
        z = TreeNode(value)
        
        #记录节点
        y = self.leaf
        x = self.root
        
        #找到插入的位置
        while x !=self.leaf:
            #记录父节点
            y = x
            if z.key<x.key:
                x= x.left
            else:
                x=x.right
        #插入动作
        z.p = y
        if y ==self.leaf:
            self.root = z
        elif z.key <y.key:
            y.left = z
        else:
            y.right = z
        z.left = self.leaf
        z.right = self.leaf
        z.color ='Red'

        self._insert_fixup(z)
    def _insert_fixup(self,z):
        #不满足的条件是父节点为红色
        while z.p.color =='Red' and z!=self.root:
            if z.p == z.p.p.left:
                y = z.p.p.right
                if y.color =='Red':
                    z.p.color ='Black'
                    y.color ='Black'
                    z.p.p.color ='Red'
                    z=z.p.p
                else:
                    if z==z.p.right:
                        z=z.p
                        self.left_rotate(z)
                    z.p.color = 'Black'
                    z.p.p.color='Red'
                    self.right_rotate(z.p.p)
            else:
                y = z.p.p.left
                if y.color == 'Red':
                    z.p.color ='Black'
                    y.color ='Black'
                    z.p.p.color ='Red'
                    z=z.p.p
                else:
                    if z==z.p.left:
                        z=z.p
                        self.right_rotate(z)
                    z.p.color = 'Black'
                    z.p.p.color='Red'
                    self.left_rotate(z.p.p)
                        
        self.root.color = 'Black'
    def Transplant(self,u,v):
        #这里没有丢掉u
        if u.p==self.leaf:
            self.root=v
        elif u==u.p.left:
            u.p.left = v
        else:
            u.p.right = v
        v.p = u.p
    def _Tree_min(self,x):
        while x.left!=self.leaf:
            x = x.left
        return  x
    def find_node(self,value):
        cur = self.root
        while cur!=self.leaf:
            if cur.key<value:
                cur=cur.right
            elif cur.key>value:
                cur=cur.left
            else:
                return cur
        return None
    def delete(self,value):
        
        node = self.find_node(value)
        if node:
            self._delete(node)
        
    def _delete(self,z):
        print(z.key)
        if not or z==self.leaf:
            return 
        y = z
        y_origial_color = y.color
        if z.left ==self.leaf:
            x= z.right
            self.Transplant(z,z.right)
        elif z.right ==self.leaf:
            x = z.left
            self.Transplant(z,z.left)
        else:
            y = self._Tree_min(z.right)
            y_origial_color = y.color
            x = y.right
            if y.p == z:
                x.p = y
            else:
                #填充y位置
                self.Transplant(y,y.right)
                #y还在，先调整他的右节点
                y.right =z.right
                y.right.p = y
            #替换节点，填充z位置
            self.Transplant(z,y)
            #再调整他的左节点
            y.left = z.left
            y.left.p = y
            #颜色不能改变
            y.color = z.color 
        
        if y_origial_color=='Black':
            self._Delete_fixup(x)
        print(y.key)
        del y
    def _Delete_fixup(self,x):
        while x !=self.root and x.color =='Black':
            if x ==x.p.left:
                w = x.p.right
                if w.color =='Red':
                    w.color = 'Black'
                    x.p.color = 'Red'
                    self.left_rotate(x.p)
                    w = x.p.right
                if w.left.color =='Black' and w.right.color =='Black':
                    w.color ='Red'
                    x = x.p
                else:
                    if w.right.color =='Black':
                        w.left.color ='Black'
                        w.color ='Red'
                        self.right_rotate(w)
                        w = x.p.right
                    w.color =x.p.color 
                    x.p.color ='Black'
                    w.right.color  = 'Black'
                    self.left_rotate(x.p)
                    x = self.root
        else:
                w =x.p.left
                if w.color =='Red':
                    w.color = 'Black'
                    x.p.color = 'Red'
                    self.right_rotate(x.p)
                    w = x.p.left
                if w.right.color =='Black' and w.left.color =='Black':
                    w.color ='Red'
                    x = x.p
                else:
                    if w.left.color =='Black':
                        w.right.color ='Black'
                        w.color ='Red'
                        self.left_rotate(w)
                        w = x.p.left
                    w.color =x.p.color 
                    x.p.color ='Black'
                    w.left.color  = 'Black'
                    self.right_rotate(x.p)
                    x = self.root
        x.color = 'Black'
    def easy_pre_sort(self,node):
        if node==self.leaf:
            return []
        return self.easy_pre_sort(node.left)+[node.key]+self.easy_pre_sort(node.right)
    def judge(self,node):
        if node==self.leaf:
            return 1,True
        if node.color=='Red':
            if node.left.color!='Black' and node.right.color!='Black':
                print('color')
                return 0,False
        if node.color not in ['Red','Black']:
            print('label',node.color)
            return 0,False
        left,l = self.judge(node.left)
        right,r = self.judge(node.right)
        if not l or not r:
            return 0,False
        if node.color=='Black':
            return left+1,left==right
        else:
            return left,left==right
 
```
